```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
housing = fetch_california_housing()
a = pd.DataFrame(housing.data ,columns = ('MedInc', 'HouseAge', 'AveRooms', 'AveBedrms',
                                                      'Population', 'AveOccup', 'Latitude', 'Longitude'))

a['target'] = housing.target
a.describe()
```

```python
a
```

```python
corr = a.corr() 
corr
```

```python
import seaborn as sb
from sklearn.preprocessing import minmax_scale
s_target = minmax_scale(a, feature_range = (0,100))
z = pd.DataFrame(s_target ,columns = ('MedInc', 'HouseAge', 'AveRooms', 'AveBedrms',
                                                      'Population', 'AveOccup', 'Latitude', 'Longitude', 'target'))
z.describe()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from matplotlib.collections import LineCollection
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from math import sqrt
from sklearn.metrics import accuracy_score,r2_score

x = z[z.columns[:8]] 
y = z.target

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = .4, random_state = 42)

reg = LinearRegression()
reg.fit(x_train,y_train)

ypred = reg.predict(x_test)
plt.xlabel('meghdar haghighi')
plt.ylabel('meghdar pish bini shode')
plt.scatter(y_test, ypred)
import sklearn.metrics 
rmse = sqrt(sklearn.metrics.mean_squared_error(y_test,ypred))
print(f'root mean square root : {rmse},\n accuracy : {r2_score(y_test, ypred)}')
```
