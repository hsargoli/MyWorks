```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
psh=plt.imshow
a=plt.imread('tes.jpg')
plt.imshow(a)
b=np.array(a)
b=b[:,:,2]
b=np.array(Image.open('tes.jpg').convert('L'))
psh(b)
```

```python
c=b.reshape(b.shape[0]*b.shape[1],1)
for i in range(c.shape[0]):
  if c[i]<np.mean(c):
    c[i]=0
  elif np.mean(c)<c[i]<100:
    c[i]=100
  elif c[i]>150:
    c[i]=200
  
d=c.reshape(b.shape[0],b.shape[1])
plt.imshow(d,cmap='gray')
```
